## NLP - Processamento de linguagem natural

* NLP é a base para mineração de texto

* Computadores estão muito longe de serem capazes de **entender** linguagem natural.

    * É preciso **escalar** computacionalmente e em **cobertura**, o que dificulta o uso de métodos muito profundos e limitados a certos domínios.

* Na prática são utilizados métodos (rasos) estatísticos de NP como base, enquanto humanos fornecem ajuda quando necessário.

### Estrutura de texto

"O cachorro está perseguindo o garoto no playground"

#### Representação em string de caracteres.

* Representa qualquer tipo de texto ou linguagem.

* Não é possível realizar reconhecimento nem ao menos as **palavras**.

* Considera todos os espaços e simbolos.

#### Segmentaçao de palavras.

* Representa mesmo texto porém em forma de sequência de palavras.

* Palavras são as **unidades básicas** da comunicação humana.

* Permitem **contagem** de palavras mais frequêntes.

* Permite formar tópicos a partir da **conexão** entre as palavras.

* Se algumas palavras são **positivas** e outras sçao **negativas** podemos realizar análise semnântica (sentimentos).

* **Generaliza menos** que a representações por caracteres.

    * Em algumas linguagens torna-se difícil **identificar** palavras.

## Recuperação da informação (RI)

Recuperação da informação é uma **sub-área** da ciência da computação e trata da recuperação **automática** da informação.

* Surgiu na **década de 60**. Principal objetivo era automatizar o acesso a informação em **bibliotecas**.

* Principais focos até o final dos **anos 80** catálogo de bibliotecas, jornais, revistas, enciclopédias **eletrônicas** e bases de dados de **empresa**.

Foco mais **recente** na **Web**:

* **Abundância** de informação não estruturada

* Publicação **sem controller** central de **diversidade**.

* Dificuldade na busca de informação **específica**, normalmente contida em **documentos**.
