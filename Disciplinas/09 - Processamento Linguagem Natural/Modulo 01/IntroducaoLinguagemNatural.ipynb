### Expressões regulares

* Exemplo de um chatbot de correspondência de padrões (comuns antes das técnicas de machine learning) usado em sistemas como Amazon Alexa.

* Um chatbot utilizando uma máquina de estados finitos que utiliza Expressões regulares para entender várias saudações.

```python
import re

r = "(hi|hello|hey)\s*([a-z]*)"
re.match(r, 'Hello Rosa', flags=re.IGNORECASE)
```

```python
re.match(r, "hi ho, hi ho, it's of to work ...", flags=re.IGNORECASE)
```

```python
re.match(r, "hey, what's up", flags=re.IGNORECASE)
```

```python
r = r"[^a-z]*([y]o|[h']?ello|ok|hey|(good[ ])?(morn[gin']{0,3}|afternoon|even[gin']{0,3}))[\s,;:]{1,3}(\w{1,20})"

re_greeting = re.compile(r, flags=re.IGNORECASE)
```

```python
re_greeting.match('Hello Rosa')
```

```python
re_greeting.match('Hello Rosa').groups()
```

```python
re_greeting.match('Good morning Rosa')
```

```python
re_greeting.match('Good morn, Rosa')
```

```python
my_names = set(['rosa', 'rose', 'chatty', 'chatbot', 'bot', 'chatterbot'])
curt_names = set(['hal', 'you', "u'"])
greeter_name = ''
match = re_greeting.match(input())

if match:
    at_name = match.groups()[-1]
    if at_name in curt_names:
        print('Good one.')
    elif at_name.lower() in my_names:
        greeter_name = at_name
        print(f'Hi {greeter_name}. How are you ?')
```

* Os primeiros trabalhos em NLP projetam manualmente regras lógicas para extrair informações de uma sequência de linguagem natural (abordagem baseada em padrões - pattern based approach)

* Estes padrões não precisam ser padrões de sequência de caracteres, como nosso exemplo de expressão regular. NLP também pode envolver padrões de sequência de palavras.

* Os primeiros componentes da PNL, como stemmers e tokenizers, bem como sofisticados mecanismos de diálogos de ponta a ponta (ELIZA¹), foram criados dessa maneira, a partir de expressões regulares e correspondência de padrões.

¹ Primeiro chatbot desenvolvido (1966). https://www.masswerk.at/elizabot/

### Processamento de linguagem natural

* Alternativa ao uso da abordagem baseada em padrões: A contrução de um banco de dados contendo sessões de diálogos entre humanos, declarações e respostas para milhoes de conversas. Seria possível ??

* Neste caso, para um chatbot, seria procurar nesse banco a mesma sequência exata de caracteres que o usuário "disse" ao chatbot.

* Problema desta abordagem: um único erro de digitação ou variação na declaração atrapalharia o chatbot.

* Para solução deste tipo de problema, podemos usar uma boardagem baseada distância (correspondeência) de caracteres:

    * Problema: medir distância entre sentenças em linguagem natural tende a obter resultados errados: sentenças similares como "bom" e "okay", podem ter diferentes caracteres (distancia medida será alta).

* No entanto, estas métricas baseadas em vetores são úteis em algumas aplicações de NLP, como corretores ortográficos e reconhecimento de nomes próprios (melhor para ortografias e não tanto para siginificados).

* Outra pssibilidade é o uso de métricas de distância entre vetores (Jaccard distance, Euclidean distance, etc).

    * Problema: pode adicionar distorçoes e atrabalhar o chatbot, quando se depara com erros ortográficos.

* Essas métricas falham em capturar a essência do relacionamento entre duas cadeias de caracteres quando são diferentes (elas tendem a dar um bom resultado quando a cadeira de caractere é muito semelhante).
