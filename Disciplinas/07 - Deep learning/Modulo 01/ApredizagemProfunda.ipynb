### Fases da pesquisa em redes neurais

* 1940 - 1960: **Cibernética**, época do "Cérebro eletrônico", de origem a moderna teoria de controle e processamento de sinais.

* 1960 - 1980: Computadores digitais, teoria de autômatos e teria de complexidade computacional. **Simples circuitos 'rasos' apresentam grandes limitações na capacidade de representação**.

* 1980 - 1990: **Cinexionismo**. Redes complexas e não-lineares, advento da propagação retrograda (back-propagation).

* 1990 - 2010: **Teoria da aprendizagem computacional**, modelos gráficos. Aprendizagem é um problema computacional dificíl, simples circuitos 'rasos' apresentam grandes limitações na capacidade de representação.

* 2006->,,,: **Aprendizagem profunda**. Treinamento de redes neurais profundas ponta-a-ponta, grandes conjuntos de dados, explosão das aplicações.

### Caracteristicas invariantes

### Quantização Vetorial

Sivic and Zisserman (2003) - "Video Google"

Adoção de tecnica de linguagem natural para processar e recuperar imagens, em particular a ideia era, dado um conjunto de imagens, descrever varios pontos de interesses e a partir dai obter, o que ficou conhecido como palavras visuais. Essas palavras poderiam ser agrupadas, e consequentemente, os descritores correspondentes de modo a constituir blocos de palavras ou descritores semelhantes e em lugar delas utilizar-se apenas um, o equivalente ao processo de exterização que realizamos no processamento de linguagem natural. Dessa forma uma imagem poderia ser representada por um histograma desses termos ou palavras visuais, e o modelo de bag of words, utilizado na recuperação de texto,  poderia ser utilizado na recuperação de imagens permitindo a verificação de semalhança a partir da ocorrencia de alguns desses termos ou palavras visuais.

### Projeto de Caracteristicas

A partir da primeira decada desse seculo, vemos a criação de varias novas caracteristicas e descritores como o Histograma de Gradiente, ou HOG, utilizado na decteção de pedestre ou modelo de parte deformaveis em regioes da imagem eram descritas por histograma de gradientes mais ainda em conjunto com o modelo de distribuição geometricas dessas regioes de forma a capturar a relação entre elas.

### Datasets desafiadores

Vemos nesse século ainda o surgimento de várias bases e datasets desafiadores, o pascal voc é uma delas. Ela continha 20 classes de objetos e detinha a tarefa de classificar esses objetos. Em 2012 temos um resultado de 80% de acuracia na identificação desses objetos. Porém alguns pesquisadores acreditavam que esse dataset tinha poucos registros.

Em 2009 em Stanford surge uma base de dados IMAGENET com 22K categorias e 14M de imagens. Por ser uma base muito grande surgiu uma nova base, derivada da IMAGENET, com 1000 clases e 1.431.167 imagens.

Entre 2010 a 2015 as abordagens mudaram.

* 2010 - Abordagem baseada em extração de caracteristica com abordagens tradicionais

* 2012 - Abordagem de aprendizagem profunda.
