```python
from bs4 import BeautifulSoup
import requests as r
```

```python
page = r.get('https://www.nytimes.com/section/science')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
from IPython.core.display import HTML, display, Image

for article in soup.find_all('article'):
    display(HTML(f'<h2>{article.div.h2.string}</h2>'))
    display(HTML(f'<p>{article.div.p.string}<p>'))
    display(Image(article.figure.a.img.get('src')))
    

    
```
