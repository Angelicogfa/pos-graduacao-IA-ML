## Formatos de conteúdo na web

### HTML

Hyper Text Markup Language Ou Linguagem de Marcação de Hipertexto.

### XML

HTML e XML são primos. Ambos identificam elementos em uma página e utilizam sintaxes similares.

A grande diferença entre HTML e XML é que o HTML descreve a **aparência** e as ações de uma **página** na rede, já o XML descreve o **conteúdo** de um documento.

### JSON

O formato JSON é eficiente e simples. Sua estrutura é de facil entendimento e utiliza convnções comuns em muitas linguagens de programação.

Muitos desenvolvedores utilizam essa alternativa, porém o XML ainda é um bom candidato, especialmente quando se quer validações mais robustas, usando os chamados XML-Schemas.

## Web Crawling vs Web Scraping

### Web Scraping

* Extração de conteúdo **específico**

Geralmente focam em sites específicos em busca de dados também especificos.

Ex: Comparação de preços, noticias de moda...

### Web Crawling

* Cobertura em **largura**

Essencialmente o que o Google, Yahoo, Bing, etc. fazem, procuram por qualquer informações e baixam conteúdos completos de paginas.

## APIs e Feeds RSS

## Web Crawling - Web Scraping

### Web Crawling

#### Coletores

Navegadores automaticos entre páginas web que visam armazenar uma **cópia** local das páginas encontradas.

Devem **obdecer** algumas restrições ao visitar sites.

O **protocolo** de exclusão de robos (Robot Exclusion Protocol) especifica algumas regras de acesso.

Principal regra é deixar um **intervalo** de tempo entre acessos a cada servidor.

**Robot.txt**

* Padrão definido em 30 de junho de 1994.

* Define as **permissões** do coletor em um determinado site.

* As diretrizes são descritas em um arquivo chamado **robot.txt**, localizado no servidor web coletado.

## Coleta de dados - API

### API - Application Programming Interface

É um conjunto definido de **mensagens** de requisição e resposta **HTTP**, geralmente expressado nos formatos **XML** ou **JSON**.

Um exemplo popular é a utilização para leitura e **publicação** de mensagens no Twitter.

No nosso caso queremos **coletar dados**.

Não é necessário a implementações de scripts externos de coleta.

É preciso criar e **registrar** uma **aplicação** para obter credenciais de acesso.

Fácil extração dos dados disponibilizados em formatos **semi-estruturados**.

### Feeds RSS

Usado principalmente em sites de notícias e blogs para compartilhamento de notícias com textos completos e até mesmos arquivos multimídias.

Um tipo de programa conhecido como "feed reader" ou **agregador** pode ler diversas páginas hablitadas para RSS.

Exemplo: agregar diversos itens relativos a carros de diversos feeds sobre automóveis.

Existem tantos sites que funcionam como feed readers como extensões para navegadores web.

#### Coleta de Feeds RSS

A coleta de conteúdo a partir de feeds rss é feita atraves de requisições HTTP.

Similiar as APIS, facilita o processo de extração dos dados, por esses já veirem em um padrão estabelecido (XML).

**Não exige autenticação** para obter os dados e tem periodicidade controlada pelo site.
