## O que é modelagem estatística.

É a parte da ciência estatística, que se ocupa de coletar, organizar, analisar e interpretar dados de determinado fenômeno (vendas, clientes,estoque, ...) que auxilie na tomada de decisões.

## Aplicações

As empresas de crescimento rápido (Startup) utilizam **modelagens estatísticas** para realizar novos **experimentos**, para testar novas ideias e validar **hipoteses**.

## Exemplo

Empresas de tecnologias no setor de multimidias (streams) criam modelos de negócios analisando o comportamento do consumidor realizando **modelagem estatistica e testes de hipoteses**.

Empresas de vendas pela internet (e-commerce) utilizam **modelagem multivariada** para predizer a compra de determinado produto.

## Conceitos

* População: Conjunto de todas as respostas, medidas ou contagens que sejam de interesse.

* Amostra: Uma parte ou subconjunto da população.

* Parametro: Descrição numérica de uma caracteristica da população.

* Estatística (estimador): Descrição numérica de uma caracteristica da amostra.

## Os dois ramos da estatistica

* Estatística descritiva: Trata da organização, resumo e organização dos dados.

* Estatística inferencial: A partir de uma amostra, tirar conclusões sobre a população.

## Tipos de variáveis

* Qualitativa Nominal: Variáveis que não são expressas em números.

* Qualitativa Ordinal: Variáveis que não são expressas em números e possuem uma ordem.

* Quantitativa Discretiva: Variáveis númericas que expressão quantidade para contagem.

* Quantitativa Contínua: Variáveis númericas que expressão quantidade no geral.

## Tipos de coleta de dados

* Experimento: Aplica-se determinado tratamento a uma parte do grupo.

* Simulação: Usa-se uma modelo matemático (em geral no computador) para reproduzir certa condição.

* Censo: Contagem ou medição de toda a população.

* Amostragem: Contagem ou medição de parte da população.

Existem **depêndencia** entre as variáveis ?

### Depêndencia (resposta vs. explicativa)

#### Resposta Quantitativa

* Regressão Linear

* Árvore de decisão

#### Resposta Qualitativa

* Regressão Logistica

### Interdependencia (Entre as variáveis exploratórias)

#### Quantitativa

* Análise Fatorial.

* Análise de Cluster.

#### Qualitativa

* Análise de Correspondência.

* Análise de Cluster.

### Interdependência (Entre as variáveis exploratórias no tempo)

#### Quantitativa

* Séries temporáris univariadas e multivariadas.
