# Séries Temporáis

Uma série temporal é um sequência de observações sobre uma variável de interesse. A variável é observada em pontos temporais discretos, usualmente equidistantes, e a análise de tal comportamento temporal envolve a descrição do processo ou fenômeno que gera a sequência.

## Padrões de Séries Temporáis

* Processo que permanecem **constantes** sobre um certo nível todo o tempo, com a variação de período a período devido a causas aleatórias.

* Padrões que ilustram **tendências** no nível dos processos de maneira que a variação de um período ao outro é atribuida a uma tendência mais uma variação aleatória.

* Processo que **variam ciclicamente (sazonal)** no tempo, como em um processo sazonal (exemplo o clima, vendas de natal).

## Modelos de Previsão de Séries Temporais

Os procedimentos de previsão de séries temporais podem ser divididos, grosseiramente, em duas categorias:

* **Automáticos**, que são aplicados diretamente, com a estilização de programas simples de computadores.

* **Não-Automáticos**, que exigem a intervenção de pessoal especializado, para serem aplicados.

### Tipos Modelos Automáticos

* Média Móvel Simples (MMS)

    * Aplicavel quando a serie tem um nível e não nem sazonalidade e tendencia.

* Alisamento Exponencial Simples (AES)

    * Aplicavel quando a serie tem um nível e não nem sazonalidade e tendencia.
 

* Alisamento Exponencial Duplo (AED)

    * Aplicavel quando tem nível e tendência, porém sem sazonalidade.

* Modelo Holt-Winters (sazonalidade) (HW)

    * Sozonalidade Aditiva

        * A Sozonalidade é constante.

    * Sozonalidade Multiplicativa

        * A sazonalidade é crescente.



| Modelo                        | Sem Sazonalidade                       | Com Sazonalidade                           |
|:------------------------------|:---------------------------------------|:-------------------------------------------|
|Sem tendencia                  |Média Movel Simples (MMS)               | Modelo Sazonal (Aditivo/Multiplicativo)    |
|                               |Alisamento Exponencial Simples (AES)    |                                            |
|Com tendencia                  |Alisamento Exponencial Duplo (AED)      | Modelo Holt-Winter (Aditivo/Multiplicativo)|



## Modelos

### Médias Móveis Simples (MMS)

Calculo da média aritimética das r últimas observações.

Previsão:

$\hat{Z_t}(h) = M_t$

$M_t = \frac{Z_t + Z_{t-1} + ... + Z_{t-r+1}}{r}$


### Alisamento Exponencial Simples (AES)

$\hat{y}_{t+1} = \alpha y_t + (1 - \alpha)\hat{y}_t$, com 0 < $\alpha$ < 1, constante de alisamento.

### Alisamento Exponencial Duplo (EAD)

$L_1 = \alpha x_1 + (1 - \alpha)(E_{t-1} + T_{t-1})$

$T_t = \beta (L_t - E_{t-1}) + (1 - \beta)T_{t-1}$

$\hat{y}_{t+k} = E_t + kT_t$

$E_t$ é a componente de nível.

$T_t$ é a componente de tendência.

$h$ é o horizonte de previsão.

$k=1,2,...,h$;

$\hat{y}_{t+k}$ é a previsão.

### Modelos Holt-Winters

Previsão de séries com tendência.

$X_t = \pi + at + S_t + \epsilon_t$

Onde: t = 1,2,....

$X_t$ é uma variável aleatória observada no tempo t.

$a$ é a tendência do modelo.

$\pi$ é o valor constante do modelo.

$S_t$ é a componente sazonal no tempo t.

$\epsilon_t$ é o erro aleatório ocorrido no tempo t.





#### Modelo Holt-Winter Simples

$L_t = a(x_t - S_{t-s}) + (1 - \alpha)(L_{t-1} + T_{t-1})$

$T_t = \beta(L_t - L_{t-1}) + (1 - \beta) T_{t-1}$

$S_t = y(X_t - L_t) + (1 - y)S_{t-s}$

$\hat{y}_{t+k} = L_t + kT_t + S_{t-s+k}$

#### Modelo Holt-Winter Multiplicativo

$L_t = \alpha \frac{X_t}{S_{t-s}} + (1 - \alpha)(L_{t-1} + T_{t-1})$

$T_t = \beta(L_t - L_{t-1}) + (1 - \beta)T_{t-1}$

$S_t = y \frac{X_t}{L_t} + (1 - y) S_{t-s}$

$\hat{y}_{t+k} = (L_t + nT_t) S_{t-s+mod(n-1,s)+1}$

### Qualidade de ajustes

* MAPE - Erro Médio Percentual

$MAPE = \frac{1}{n} \sum_{i=1}^n |{\frac{y_t - \hat{y}_t}{y_t}}| x 100$

* MAD - Erro Médio Absoluto

$MAD = \frac{1}{n} \sum_{i=1}^n |y_t - \hat{y}_t|$

* MSD - Erro Quadrático Médio

$MSD = \frac{1}{n} \sum_{i=1}^n (y_t - \hat{y}_t)^2$
