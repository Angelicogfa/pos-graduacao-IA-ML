# Análise de Imagens e Visão Computacional

## Introdução a visão Computacional

* Um problema de visão computacional é usar uma imagem para inferir algo sobre o mundo.

* O objetivo é "entender" o conteúdo visual de uma imagem.

* Uma imagem é definido por pixels que são a menor unidade de informação em uma imagem.

* Uma imagem fullhd tem 1080 x 1920 px. = 2.073.600 px

* Um histograma de cor é a distribuição de todas as cores entre os pixels.

* Uma escala de cinza a escala de cor vai de 0 a 255 em um valor de 8 bits

* Uma imagem colorida tem um pixel com 3 valores, referente aos 3 canais de cores do padrão RGB, as tonalidades vão de 0 a 255 no três canais.
