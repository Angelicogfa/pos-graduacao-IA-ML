## Deep Learning

A ideia do aprendizado profundo (deep) est relacionada sucessivas camadas (layers) de representao.

Estas camandas de representao so construdas por meio de modelos de redes neurais. Assim, **deep learning  um framework para aprendizado de representao a partir de dados** (input).

```python
display(Image('Imagens/Imagem0004.png'), Image('Imagens/Imagem0005.png'))
```

A transformao implementada pela camada  parametrizada pelos seus pesos (parmetros de entrada da camada), que so iniciados aleatoriamente.

"Aprender" siginifica encontrar um conjunto de valores para os pesos de todas as camadas da rede, de forma que a rede ir mapear corretamente exemplos de entrada para a sua sada.

Para medir a sada, devemos saber o quo distante a sada esta do resultado esperado (loss function).

A *loss function* calcula o quo distante a predio est do resultado esperado.

Assim, o score  obtido como o feedback para ajustar o valor dos pesos da rede, com o intuito de diminuir o valor obtido pela *loss functions* *(optimizer)*.

O processo  chamado **backpropagation**.

No incio, como os pesos so iniciados aleatoriamente, o valor da loss  muito alto.

Com os ajustes dos pesos de pouco em pouco, a loss tende a cair.

## Anatomia de uma rede neural

* Layer (camadas): so combinadas dentro de uma rede (modelo)

* Loss Function: define o valor de feedback usado para o aprendizado.

* Optimizer: determina como o aprendizado ocorre.

### Como uma rede manipula um vetor ??

Por meio de um tensor 2D (um array de vetores) onde o primeiro axis (nome da primeira dimenso de um tensor)  chamado de *sample axis* e o segundo axis  chamado de *feature axis*.

Por exemplo, um dataset de pessoas com idade, cpf e nome. Cada pessoa seria representada com um vetor de 3 valores. Para um dataset de 100.000 pessoas, um tensor teria a seguinte forma $(100000, 3)$

### Camadas e contruo dos blocos

* Layers (camadas)  a mais fundamental estrutura de dados de uma rede.

* Recebe como entrada um ou mais tensores e retorna um ou mais tensores.

* Diferentes tipos de camadas so usadas para diferentes tipos de operao.

* Para um vetor, armazenado em um tensor 2D, normalmente utiliza-se uma **fully connected** or **dense layers** (Dense no keras)

* Imagens normalmente so processadas por uma **2D convolutions layers** (Conv2D no keras).

### Desenvolvimento com Keras

Tpico keras workflow:

1. Define dataset de treinamento: input tensor e target tensor

2. Define a network layer (model) que mapeia os inputs para os targets

3. Configura o processo de aprendizado escolhendo a loss function, um optimizer, e algumas metricas para monitoramento.

4. Iterao com o dataset de treinamento chamando o mtodo fit() do seu modelo
