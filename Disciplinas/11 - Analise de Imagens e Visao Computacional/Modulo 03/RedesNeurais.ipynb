## Deep Learning

A ideia do aprendizado profundo (deep) está relacionada sucessivas camadas (layers) de representação.

Estas camandas de representação são construídas por meio de modelos de redes neurais. Assim, **deep learning é um framework para aprendizado de representação a partir de dados** (input).

```python
display(Image('Imagens/Imagem0004.png'), Image('Imagens/Imagem0005.png'))
```

A transformação implementada pela camada é parametrizada pelos seus pesos (parâmetros de entrada da camada), que são iniciados aleatoriamente.

"Aprender" siginifica encontrar um conjunto de valores para os pesos de todas as camadas da rede, de forma que a rede irá mapear corretamente exemplos de entrada para a sua saída.

Para medir a saída, devemos saber o quão distante a saída esta do resultado esperado (loss function).

A *loss function* calcula o quão distante a predição está do resultado esperado.

Assim, o score é obtido como o feedback para ajustar o valor dos pesos da rede, com o intuito de diminuir o valor obtido pela *loss functions* *(optimizer)*.

O processo é chamado **backpropagation**.

No início, como os pesos são iniciados aleatoriamente, o valor da loss é muito alto.

Com os ajustes dos pesos de pouco em pouco, a loss tende a cair.

## Anatomia de uma rede neural

* Layer (camadas): são combinadas dentro de uma rede (modelo)

* Loss Function: define o valor de feedback usado para o aprendizado.

* Optimizer: determina como o aprendizado ocorre.

### Como uma rede manipula um vetor ??

Por meio de um tensor 2D (um array de vetores) onde o primeiro axis (nome da primeira dimensão de um tensor) é chamado de *sample axis* e o segundo axis é chamado de *feature axis*.

Por exemplo, um dataset de pessoas com idade, cpf e nome. Cada pessoa seria representada com um vetor de 3 valores. Para um dataset de 100.000 pessoas, um tensor teria a seguinte forma $(100000, 3)$

### Camadas e contrução dos blocos

* Layers (camadas) é a mais fundamental estrutura de dados de uma rede.

* Recebe como entrada um ou mais tensores e retorna um ou mais tensores.

* Diferentes tipos de camadas são usadas para diferentes tipos de operação.

* Para um vetor, armazenado em um tensor 2D, normalmente utiliza-se uma **fully connected** or **dense layers** (Dense no keras)

* Imagens normalmente são processadas por uma **2D convolutions layers** (Conv2D no keras).

### Introdução ao Keras

* Keras é um framework para deep learning em Python

* Keras é uma biblioteca alto nível. Keras não manipula operação baixo nível como operação sobre tensor.

* Keras utiliza um backend engine: TensorFlow backend, Theadno backend, Microsoft Cognite Toolkit (CNTK)

### Desenvolvimento com Keras

Típico keras workflow:

1. Define dataset de treinamento: input tensor e target tensor

2. Define a network layer (model) que mapeia os inputs para os targets

3. Configura o processo de aprendizado escolhendo a loss function, um optimizer, e algumas metricas para monitoramento.

4. Iteração com o dataset de treinamento chamando o método fit() do seu modelo
