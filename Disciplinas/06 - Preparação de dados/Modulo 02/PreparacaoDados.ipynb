## Tipos conceituais de variavies

* Variavel escalar: Valor da média+escaça da medida. (Exemplo velocidade de 50 m/s). Podem ser quanittativas ou valitativas.

* Variaveis vetoriais: Precisa de mais de um valor para defini-la. (Exemplo: vetor de localização [10' latitude, 30' longitude]). Cada valor possui sentido individual em conjunto.

## Variaveis Quantitativas

* Variaveis discretas: Podem assumir apenas um número finito ou infinito contável de valores entre dois valores. Correspondem a dados inteiros. Exemplos: Número de filhos, número de vezes que exercita por mês, número de bacterias no leite. etc.

* Variáveis continuas: Caracteristicas que assumem valores infinitos entre dois valores numa escala contínua, onde valores fracionais são relevantes. Exemplo: preso, altura, idade, tempo, etc.

## Variáveis Qualitativas ou categoricas Politómicas.

* Variáveis nominais: Não existem uma ordem natural entre as categorias. Exemplo: Sexo, estado civil, religião, profissão, raça, cor de pelo. etc.

* Variáveis ordinais: Existe uma ordenação natural entre as categorias. Exemplo: nível de escolaridade, mês de observação.

## Variaveis Qualitativas ou Categoricas Dicotómicas

* Variaveis que podem adotar somente dois valores. Exemplo: sexo (M/F).

## Variáveis qualitativas nominais

* Refere-se aos nomes das "coisas" que nominalmente é impostos por nós.

* Exemplos: 

    * Alimentos: café, azeitona, palmito.

    * Vestimenta: calça, camisa, blusa.

    * Artigos de escritório: caneta, papel, lápis.

* Essas medidas não possuem uma ordem inerente, poderão ser codificadas de forma arbitrária como (1,2,3), (A,B,C) ou plelos próprios nomes atribuidos dos valores.

* Cada medida (nome) possui diferenças claras na sua identidade, mas essas diferenças não pode ser manipulada matematicamente. A sequência dos npumeros é somente uma conveniência, não implicam numa ordem que quantifique a diferença.

## Variável qualitativa ordinal

"Pense na medicação da temperatura do se seu café."

* A medição pode estar limitada a "quente" ou "frio".

* A medição anterior possui menos informação que "fervendo", "muito quente", "quente", "morno", "frio", "congelado".

* Dependendo da aplicação pode ser necessário aumentar ou diminuir o tamanho do grão. Grãos maiores proporcionam resultados mais especificos.

* Para as medidas pertencerem a uma escala ordinal é necessário cumprir a condição de TRANSITIVIDADE.

Se (A > B) e (B > C) então (A > C)

* Caso seja necessário criar uma nova codificação para a variável ordinal é necessário cumprir a condição de TRANSITIVIDADE.

```python
import numpy as np

idades = np.array([49, 63, 44, 25, 16,34, 62, 55, 40, 31, 44, 37, 48, 65, 83, 53, 39, 15, 25, 52, 68, 35, 64, 71, 43, 76, 39, 61,51,30,32,74, 28, 64, 46, 31, 79, 69, 38, 69, 53, 32, 69, 39, 32, 67, 17, 52, 64, 64, 25, 28, 64, 65, 70 ,44 ,43, 72, 37, 31, 67, 69, 64, 74, 32, 25, 65, 39, 75, 36, 26, 59, 28, 23, 40, 56, 77, 68, 46, 48])
```

```python
idades.sort()

# Obtendo as variaveis para calculo da distribuição
menor = np.round(idades.min(), 1)
maior = np.round(idades.max(), 1)

# Calculando os indices para montagem da tabela
n = len(idades)
a = maior - menor
k = (1 + 3.322 * np.log10(n)) // 1
ak = np.round(a / k, 1)

# Definindo as frequencias
frequencias = np.array([])

valor = menor
while valor < idades.max():
    nvalor = np.round(valor + ak, 1)
    frequencias = np.append(frequencias, f'{valor}|--{nvalor}')
    valor = nvalor

# Pesquisando os registros com base nas frequencias definidas
freq_abs = pd.cut(idades, len(frequencias), labels=frequencias)

# Montando o dataframe
df = pd.DataFrame(freq_abs.value_counts(), columns=["Frequência"])
df.reset_index(inplace=True)
df.rename(columns={"index": "Idades"}, inplace=True)
df["Percentual"] = df["Frequência"] / n * 100
df["Frequência Acumulada"] = df["Frequência"].cumsum()
df["Percentual Acumulado"] = df["Percentual"].cumsum()
df
```

### Estatistica descritiva de variáveis - Histograma de frequências

```python
plt.title('Frequência das idades')
plt.barh(df['Idades'], df['Frequência'])
plt.xlabel('Frequência')
plt.ylabel('Idades')
plt.show()
```

### Distribuição Normal

A distribuição normal é um modelo estatístico que fornece uma base teórica para o estudo do padrão de ocorrência dos elementos de uma população.

$f(x) = \frac{1}{\sqrt{2 \pi \sigma}} exp \lbrace - \frac{(x-\mu)^2}{2\sigma^ 2} \rbrace$

$f(x)$ média da população (ou da amostra).

$\sigma (x)$ desvio padrão da população (ou da amostra).
