# Visualização de dados

## O que é visualização de dados ?

* Muito provavelmente a primeira coisa que veio em nossa mente é:
    * Graficos
    * Dashboards

* Visualização de dados é algo muito além de gráficos e ou Dashboards.

* É o estudo e a criação de representações visuais para dados.

## Representações visuais

* Você não precisa fazer uso de gráficos em ferramentas ligadas a tecnologia para representar seus dados.

## Interpretação

* Gráficos e textos são **linguagens informais**

* Dependenm de contexto, background cultural.

* Para serem úteis precisam ser interpretados, aprendidos, transformados em informações e conhecimento.

* Interpretações frequentemente são conflitantes.

## Visão é convenção

* Representaçãoes visuais são **aprendidas**; percebemos mais facilmente o que já conhecemos.

* Podemos entender gráficos complexos e compactos, mas podemos no enganar na primeira impressão.

* Objetivo da visualização é encontrar maneiras mais **simples** e **precisa** para compreender e comparar dados: transformá-los em informação e conhecimento.

## Revelar o invisível

* A visualização de dados permite trabalhar com universos e dimenões que não existem no mundo das coisas visíveis.

* Espectros de luz invisível

* Quatro ou mais dimensões físicas

* Ondas e linguagens sonoras

* O infinitamente grande e o infinitamente pequeno

* Abstrações matemáticas e filosóficas, conceitos e idéias, quantidades, estatísticas.

## Por que visualizar ?

* Para ter uma visão abrangente dos dados (é dificil lembrar de muitos números em uma tabela)

* Para comparar valores rapidamente

* Para descobrir padrões dentre os valores

* Para comparar padrões

## Por que gráfico ?

"O poder de um gráfico é permitir que informações quantitativas seja organizada de forma a **revelar** padrões e estruturas que não são facilment detectadas por outros meios de análise dos dados" (Cleveland & McGill).

## As representações mais eficientes

* Mais simples (rapidamente compreendido)

* Mais atraente (prendem a atenção)

* Seguem convenções (usam formas já aprendidas)

* Mais fácies de comparar (posição, comprimentos)

* Mais compactados (pontos mais próximos são mais fáceis de comparar)

## Não deixe de olhar o seu publico

2. Posições em escalas não alinhadas

3. Comprimento, direção, ângulo

4. Área

5. Volume, curvatura

6. Sombras, saturação de cor

## continuação...

* A importância da **precisão** é maior quanto mais analítica ou exploratória for a visualização.

* Nos outros casos é possível escolher o nível ideal de compromisso entre precisão e interesse visual necessária para a **história** que esta sendo contada a seu **público**.

## Como escolher a melhor forma de visualização ?

### Percepção

* Barras (comprimento), pontos e linhas (posição) são mais facilmente percebidos

* Ângulos e áreas são mais difíceis de comparar

* Saturação e tonalidade não são adequados para representar dados quantitativos.

### Antes de começar

* Organize seus dados.

* Saber sua origem e quais outras fontes serão cruzadas, quando for o caso.

* Estruturar de acordo com uma pré-visualização (protótipo).

* Contar história sobre os dados.
