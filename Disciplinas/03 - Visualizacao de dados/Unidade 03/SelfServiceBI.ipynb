# Self-Service BI

## O que é ?

É a democratização dos dados. Ele permite que os usuários de negócios (como executivos ou gestores) acessem visões a partir dos dados disponíveis sem a assistência de TI, ou uma equipe de analistas.

## Na literatura

Imhoff e White (2011) reportam que as constantes mudanças nos negócios, incapacitade da TI de atender a grande demanda de informações de negócios e necessidade de informações mais analíticas são os principais fatores que fazem com que as empresas busquem alternativas ao tradicional BI. (ssbi-jul12-11.pdf)

## Na web

* Segundo a Gartner (2018) Self-Service de Business Intelligence pode ser definido onde os usuários finais podem criar seus próprios relatórios, dentro de um portifólio de arquitetura e ferramentas aprovadas.

* Segundo BI9 (2019) o conceito **Self-Service BI** tem sido muito falado nos últimos tempos e não é para menos, a ideia de ter autonomia para analisar dados sem precisar da intervenção do setor de TI, permite que o usuário possa obter respostas e tomar decisões com muito mais agilidade.

## Onde se aplica

* São ferramentas que possibilitam através de conectores, acessar e combinar diversas forntes de dados.

* Sua aplicação principal está na possibilitadade de fazer diversas análises com vários recursos visuais.

## Possibilidades

* Compartilhamento imediato das informações

* Maior dependência da TI

* Disponibilidade de informações dinâmicas e interativas, permitindo que o usuário final crie/modifique seus relatórios.

* Relatórios interatvivos em tempo real possibilitando visualizações, pesquisas dinâmicas e livre exploração de dados, facilitando a análise e impactando diretamente na resolução de problemas críticos ou operacionais.

## Porque adotar SSBI

* Usuários desenvolvem no seu desktop e depois tem a possibilidade de compartilhar seus dashboards pela web e muitas vezes na nuvem ou ainda através de dispositivos móveis.

* Possui interface bastante amigável e intuitiva que "favorece" a compreensão das informações, obviamente, se a história estiver bem contada.
