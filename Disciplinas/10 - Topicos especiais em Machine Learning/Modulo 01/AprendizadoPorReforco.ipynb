## Aplicações

* Andar de robos

## Conceitos

* Área de aprendizado de máquina que lida com decisões sequenciais

* Construido por um agente tomador de decisões (toma ação) e um mundo em que o agente interage, chamado ambiente.

* A meta do agente é maximar o total de recompensas acumuladas.

* Ele busca otimizar o resultado a cada ação.

* Uma recompensa $R_t$, é um sinal de feedback escalar.

* Indica quão bem o agente está no espaço de tempo $t$.

Aprendizado por reforço é baseado na hipótese de recompensa.

## Definição (Hipótese de Recompensa)

Todas as metas podem ser descritas pela maximização da esperança da recompensa acumulada.

## Exemplo de recompensa

* Acrobacia de voo de helicopetero

    * (+) voar na trajetoria correta

    * (-) bater ou cair

* Fazer locomoção de humanoide

    * (+) Movimentos para frente

    * (-) cair

* Jogar atari melhor do que humanos

    * (+) aumentar pontuação

    * (-) diminuir pontuação

## Tomada de decisão sequencial

* Meta: maximizar a recompensa total futura

* As ações podem ter consequências de longo prazo

* Recompensa pode ser atrasada

* Talvez seja melhor sacrificar recompensas imediatas para ganhar recompensas de longo prazo

* Exemplo:

    * Investimentos financeiros

    * Reabasteciemento de helicóptero

## Histórico e estado

* Um histórico é uma sequência de observações, ações e recompensas $H_t = O_1, R_1, A_1, ..., A_{t1}, O_t, R_t$

* Por exemplo: Todas as variáveis até o tempo $t$

* O que acontece no próximo instante depende do histórico:

    * O agente seleciona ações

    * O ambiente seleciona observações/recompensas

* Estado e a informação usada para determinadr o que acontece no próximo intervalo de tempo $S_t = f(H_t)$

* Um estado de informação é o conhecimento como **estado de Markov**, ele contém toda informação útil do histórico.

### Definição:

Um estado S, é Markov se, e somente se:

$P[S_t+1|S_t] = P[S_t+1|S_1, ..., S_t]$

* O futuro é independente do passado dado o presente

$H_{1:t} \overrightarrow{} S_t \overrightarrow{} H_{t+1:\infty}$

## Componentes de um agente

Um agente pode conter um ou mais destes componentes:

* Politica: função de comportamento do agente

* Função valor: quão bom é o estado e/ou ação

* Modelo: representação do ambiente do agente

### Politica

* É o comportamento do agente

* É um mapeamento do estado para ação, por exemplo.

* Determinístico: $a = \pi(s)$

* Estocástico: $\pi(a|s) = P[A_t = a|S_t = s]$

### Função valor

* A predição do valor de recompensa futura

* Usada para avaliar quão bom/ruim cada estado É

* Assim, seleciona ações, por exemplo.

$v_{\pi}(s) = E_{\pi}[R_t + \gamma{R_{t+1}} + \gamma^2R_{t+2} + ... |S_t = s]$

### Modulo

* Prediz o que o ambiente fará no próximo espaço de tempo.

## Introdução - Processo de decisão de Markov (MDPs)

* Formalmente descreve um ambiente de aprendizado por reforço

* Ambiente totalmente observável

* Quase todos problemas de aprendizado por reforço podem ser caracterizados como MDPs.


### Propriedade de Markov

Definição: Um estado $S_t$ é Markov se, e somente se:

$P[S_t+1|S_t] = P[S_t+1|S_1, ..., S_t]$

* O futuro é independente do passado dado o presente.

### Processo de Markov

Processo randômico sem memória, como uma sequência de estados randômicos $S_1, S_2, ...$ com a propriedade de Markov.


Definição:

Um processo de Markov ou uma cadeia de Markov é uma tupla $\{S,P\}$

* $S$ é um conjunto finito de estados

* $P$ é uma matriz de probabilidades de transição de estados

    $P_{ss'} = P[S_{t+1} = s'|S_t = s]$

## Processo de Recompensa de Markov (MRP)

Definição

Um processo de Markov ou uma cadeia de Markov é uma tupla $ \{
S, P, R, \gamma \}$

* $S$ é um conjunto finito de estados

* $P$ é uma matriz de probabilidade de transição de estados

    $P_{ss'} = P[S_{t+1}  = s'|S_t = s]$

* $R$ é uma função de recompensa

    $R_s = E[R_{t+1}|S_t = s]$

* $\gamma$ é um fator de desconto $\gamma \in [ 0, 1 ]$

## Retorno

Definição:

O retorno $G_t$ é a recompensa total descontada no espaço de tempo $t$.

$G_t = R_{t+1} + \gamma R_{t+2} + ... = \sum^{\infty}_{k=0} \gamma^k R_{t+k+1}$

## Função Valor no MRP

Representa o valor de recompensa de longo prazo a partir de $S$

Definição:

É o retorno esperado iniciando do estado $s$.

$v(s) = E[G_t|S_t = s]$
