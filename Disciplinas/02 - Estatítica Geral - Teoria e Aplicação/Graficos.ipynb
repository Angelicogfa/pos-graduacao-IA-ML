# Graficos

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
plt.style.use('ggplot')
```

## Grafico de Setores

* É um grafico muito comum para representar distribuições de frequencia qualitativas
* É particularmente util quando o número de categorias não é grande e as caracteristicas não obedecem a alguma ordem especifica
* **Vantagem**: Todas as informações na tabela de frequencia podem ser transposta para o gráfico.
* **Desvantagem**: Não suporta muito tipo de categorias.


```python
# Função para gerar a legenda do gráfico
def gerar_legenda(pct, allvals):
    absolute = int(pct/100 * np.sum(allvals))
    return f'{pct:.1f}%\n({absolute} kg)'

# Criando a área de plotagem do gráfico
fig, ax = plt.subplots(figsize = (6,6), subplot_kw=dict(aspect='equal'))

# Dados do grafico
labels = ['Dentro das especificações', 'Fora das especificações']
values = [320, 47]

# Criando o grafico
wedges, texts, autotexts = ax.pie(values, autopct= lambda pct: gerar_legenda(pct, values), textprops=dict(color='w'))

# Criando a caixa de legenda externa
ax.legend(wedges, labels, title='Especificações', loc='center left', bbox_to_anchor=(1,0,0.5,1))

# Definimos o tamanho da fonte
plt.setp(autotexts, size=8, weight='bold')

# Definimos o titulo do grafico
ax.set_title('Distribuição de peças produzidas')

plt.show()
```

## Gráfico de barras ou colunas

* Utiliza o plano cartesiano com os **valores ou categorias da variável no eixo das abcissas (x)** e as **frequencias ou porcentagens no eixo das ordenadas (y)**.
* Para cada valor ou categoria da variável desenha-se uma barra com altura correspondendo à sua frequencia ou porcentagem.
* Esse tipo de gráfico se adapta bem à **variáveis qualitativas (ou categorias)** ou **quantitativas discretas (valores numericos inteiros - conta-se para chegar ao valor. Ex: número de alunos, número de acidentes)**
* **Vantagem**: Comporta uma quantidade bem grande de valores e categorias.
* **Desvantagem**: Não consigo explicitar percentual e frequencia ao mesmo tempo, pelo motivo do grafico ser plotado sobre um plano carteriano.

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 17, 7, 6]

# Area de plotagem do grafico
fig, ax = plt.subplots(figsize = (8,6))

rect = ax.bar(labels, values, color='green')
ax.set_ylim([0,20])
ax.set_xlabel('Provedor')
ax.set_ylabel('Número de visitantes')
ax.set_title('Distribuição do provedor pelo visitante de um determinado site', pad=25)

def criar_label(rects):
    # Itera Pelos retangulos do grafico
    for rect in rects:
        # pega a altura de cada retangulo
        height = rect.get_height()
        # Cria uma anotação 
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rect)

fig.tight_layout()
plt.show()
```

## Grafico de linhas



* É particularmente indicado para retratar **séries temporais**, ou seja, d*ados relativos a variaveís observadas anualmente, mensalmente, trimestralmente, de hora em hora, diariamente, etc*..
* Sua construção é muito semelhante à do gráfico de barras porém utiliza-se linhas interligadas ao invés de barras para representar a frequência ou a porcentagem de cada valor ou categoria da variável estudada.


```python
dias = np.asarray(range(1,35,1))
manutencoes = [18,22,27,24,33,24,23,32,12,38,12,35,24,24,27,26,20,22,12,20,29,27,25,38,27,24,23,19,18,16,23,25,22,31]

fig, ax = plt.subplots(figsize=(8,8))

ax.plot(dias, manutencoes, color='gray', marker='o', linewidth=2, markersize=6,  markerfacecolor='red')
ax.set_xlim([0,35])
ax.set_ylim([0,40])
ax.set_ylabel('Nº de manutenções')
ax.set_xlabel('Dia')
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
ax.set_title('Número de matenuções diárias em equipamentos durante 5 semanas', pad = 25)
plt.tight_layout()
plt.plot()
```
