# Arvores de decisão

## Princípio

O algoritmo de arvóre de decisão segue principios básicos, sua estrutura é segue é baseada em nos de dados, onde existe um nó de entrada inicial, que por sua vez possui nós filhos e os nós finais, das pontas, são chamados nós folhas.

* Cada nó é um atributo base de dados.

* Nós folhas são artibutos-classe (rótulo/label)

* Cada ramo ligando um nó-filho a um nó pai é etiqetado com o valor do atributo do nó-pai.

* Um atributo que aparece num nó não pode aparecer em seus nós descendentes.

### Indução da árvore

* ID3 (final dos anos 1970) - iteratuve duchotomiser

    * Só aceita dados categoricos.

    * Nao funciona muito bem com dados omissos e outliers.
    
* C45 (sucessor do ID3)

    * A criação da arvóre é binária

* CART (1984) - Classification and Regression Trees

    * Utiliza o Jine (escola de atributo)

* J48

    * Implementação do C45 em java.

### Tipos de dados

* ID3: dados categóricos

* C4.5: dados contínuos, suporta dados omissos.

### Parâmetros de entrada

* Base de conhecimento (B).

* Lista de atributos candidatos (CAND).

* Um atributo-classe (rótulo): sempre categórico.

### Método de selação de atributos (principais formas)

* Ganho de informação (ID3).

    * Baseado na entropia dos dados (variação)

    * Muito sensível a outliers.

* Taxa de ganho (C4.5, J48).

    * Razão com base na entropia.

* Índice GINI - impureza (CART).



### Visão geral do algoritmo de ID3 (C4.5)

1. Cria um nó $N$ associado à base de dados base $B$

* Se todos os registros de $B$ pertencem à mesma classe $C$, então transforme $C$ em um nó folha rotulado por $C$.

* Se não, se $CAND = ()$, então transforme $N$ numa folha etiquetada com o valor $C = max(count(atributo-classe(A)))$

* Se não, seleciona atributo-teste $A = max(Ganho(CAND))$ e rotule $N$ com o nome de atriutos-teste $A$

2. Partição das amostras de $B$

* Para cada valor $s_i$ do atributo-teste, faça

* Crie um nó-filho $N$ ligado a $N$ por um ramu rotulado pelo valor $s_i$, e associe a este nó uma sub-classe $B$, tal que o $atributo-teste = s_i$

* Se $B = ()$, então transforme o $N_i$ numa folha etiquetada com o valor $C = max(count(atributo-Classe(A)))$

* Se não, calcule $Arvore(B_i, CAND - (atribut-teste))$ e associe ao nó $N$.

### Seleção de atributos: Ganho de informação

* Dados categóricos (npumero de categorias = v)

* Entropia: $E(p, n) = - \frac{p}{p+n} log_2 \frac{p}{p+n} - \frac{n}{p+n} log_2 \frac{n}{p+n}$

* Usando o atributo $A$, a base de dados $B$ será particionada em conjuntos $S$. A quantidade de informação finali será:

    * $I(A) = \sum^{v}_{i=1} \frac{p_i+n_i}{p+n} I (p_i, n_i)$

* Ganho de informação: $G(A) = I(p,n) - I(A)$

### Selação de atributos: Taxa de Ganho

* Dados categoricos ou contínuos (C4.5)

* Usado para dados categóricos com muitos valores:

    * Ganho da informação tende a produzir overfiting nesses casos.

### Seleção de atributos: Índice Gini

* Dados contínuos (IBM Intelligent Miner)

* Se uma baes $B$ contém amostras de $N$ classes:

    * $gini(B) = 1 - \sum^{n}_{j=i} p j^2$

    onde $pj$ é a frequência relativa da classe $j$ em $B$.

* Se $B$ é particionada nas subclasses $B1$ e $B2$ com tamnhos $N_1$ e $N_2$:

    * $gini_{part}(B) = \frac{N_1}{N} gini(B_1) + \frac{N_2}{N} gini(B_2$)

Considere uma hipotese *h*:

Diz-se que *h* sofre overfiting referente aos dados de treinamento se:
