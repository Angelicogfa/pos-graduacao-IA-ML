### Variáveis Contínuas

Qualquer distância métrica pode ser utilizada.

Mais importantes são classes de distância de Minkowski:

$d(i, j) = \sqrt[n]{(|x_{i1} - x_{j1}|^n + |x_{i2} - x_{j2}|^n + ... + |x_{ip} - x_{jp}|^n)}$

* se n = 1, d é a distância de Manhattan.

* se n = 2, d é a distância Euclidiana.

### Z-score

$Z = \frac{x - \mu}{\sigma}$

* x: valor; $\mu$: média; $\sigma$: desvio padrão.

* Distância entre o dado e a população em termos do desvio padrão.

* Negativo quando abaixa da média, e positivo caso acima.

### Medidas de similaridade baseadas em vetor

Em alguns casos, medidas de distância provêm visão distorcida.

* Ex: Quando o dado é muito esparso e 0's no vetor não são significativos.

* Nesses casos, melhor utilizar medidas de distância baseadas em vetor.

$X = (x_1, x_2, ..., x_n)$
$Y = (y_1, y_2, ... =, y_n)$

#### Medida de similaridade de cosseno

* Produto escalar: $sim(X, y) = X * Y = \sum x_i \ * \ y_i$

* A norma do vetor X é: $\|X\| = \sqrt{\sum x^2_ i}$

* A similaridade de cosseno é:

$sim(X, Y) = \frac{X * Y}{\|X\| * \|Y\|} = \frac{\sum x_i \ * \  y_i}{\sqrt{\sum x^2_i} * \sqrt{\sum y^2_i}}$

#### Exemplo:

X = <2,0,3,2,1,4>

$\|X\| = \sqrt{(2^2+0^2+3^2+2^2+1^2+2^2)}  = \sqrt{(4+0+9+4+1+16)} = 5.83$

$X* = \frac{X}{\|X\|}$ = <0.343, 0.0514, 0.343, 0.171, 0.686>

* Note que  $\|X\| = 1$

* Torna o vetor de comprimento unitario.

#### Exemplo de similaridade entre documentos

## Métodos baseados em particionamento

Dada uma base de dados de $n$ elementos e um número de clusters $K <= n$:

* Cria-se uma participação inicial aleatória de $k$ partes

* Num processo iterativo, os elementos das partes são realocados para outras partes de tal modo a melhorar o particionamento.


### Adequado para descobrir cluster de formato arbitrário

* Clusters são regiões densas de objetos no espaço de dados separadas por regiões de baixa densidade (representando ruídos).

* Região densa possui uma x-vizinhança de cada ponto (onde x é um parâmetro dado) que contém pelo menos x pontos.

## Métodos Hierárquicos aglomerativos

* Inicialmente, cada elemento da base forma um cluster.

* A cada iteração pares de clusters mais próximos são aglutinados num único cluster.

* Termina quando número de clusters K é atingido.

* Exemplo: AGNES (AGlomerative NESting).

## Métdos Hierárquicos divisórios

* Inicialmente, cria-se um único cluster composto por toda a base.

* A cada iteração os clusters são subdivididos em duas partes.

* Determina quando número de clusterks k é atingido.

* Exemplo: DIANA (Divisive ANAlysis).
