# Aprendizado Supervisionado

Aprendizado supervisionado é um ramo do aprendizado de máquina onde um modelo é treinado com base em instâncias de registros já classificados anteriormente, ou seja, o modelo aprende as possíveis classificações da base já roladas.


## Separação das bases de dados

A separação da base de dados é um importante passo pois é nesse momento que separamos o conjunto de dados em duas amostras com tamanhos diferentes. Esse tamanho precisa ser bem definido por que pode acontecer das bases (treino e teste) não possuirem todo o espaço amostral de possibilidades, assim deixando o modelo menos propenso a generalização, pois aquele tipo de instancia não será conhecida.

### Base de treinamento e de testes

* Devem ser semelhantes (estatisticamente, cobertura do espaço de solução, etc.)

* Se a  base de dados é grande, pode-se partir a base (pergentage split).

* Caso contrário, usar validação cruzada (cross validation)

* Percentage Split/Houd out

    * Particiona a base por amostragem

    * Tipos de amostragem inferem no resultado

* Cross validation

    * Particiona em K-partes

    * Exemplo: Separa a base em 10 partes: Em cada rodada usa 9 blocos para treinamento e 1 para validação

### Seleção de atributos

A seleção dos atributos, ou caracteristicas, é uma tarefa crucial para a geração de um modelo que atenda todos os requisitos desejado pelo Cientista de dados. Dependendo da quantidade de caracteristicas ou da qualidade, o modelo pode ser impactado negativamente por algumas questões.

* Feature selection / feature engineering.

* Normalmente redundância não é problema, mas afeta alguns algoritmos.

* Agregar quando for relevante.

* Pode-se usar redução de dimensionalidade.

* Afeta o desempenho.

O PCA (analise de componentes principais) é um recurso bastante utilizado em processos de redução de dimensionalidade, principalmente nos processos de analise de imagens, onde um vetor terá suas dimensões reduzidas mas sem perder as caracteristicas fundamentais.



### Escolha do algoritmo

A escolha do algoritmo deve levar em consideração varios aspectos existentes na base de dados que podem impactar diretamente na perfornance do modelo.

* Tarefa: classificação ou previsão ?

* Tipos de dados

* Distribuição das classes: são balanceadas ? 

* Interpretação dos resultados.
