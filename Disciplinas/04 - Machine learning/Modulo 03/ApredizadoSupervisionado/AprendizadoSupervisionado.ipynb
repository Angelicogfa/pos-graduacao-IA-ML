A separação da base de dados é um importante passo pois é nesse momento que separamos o conjunto de dados em duas amostras com tamanhos diferentes. * Caso contrário, usar validação cruzada (cross validation)

* Percentage Split/Houd out

* Cross validation

* Feature selection / feature engineering.

### Avaliação do modelo: classificação binária

*Matriz de confusão*

* Mostra o número de classficações corretas versus as classificações preditas para cada classe, sobre um conjunto de exemplo T.

* A matriz de confusão de um classificador ideal possui valores não nulos apenas na diagonal

* *Previsão*: Acertar corretamente a classificação

* *Revocação*: Acertividade do modelo com relação a recuperar a classificação da classe A 

*ex*:

Em uma triagem de um hospital, todos os pacientes que chegam tem sua pressão, temperatura, altura e peso afetidas e idade e sexo identificados a fim de determinar se o paciente possui um risco cardiaco. Dado a situação:

Temos que:

* A *precisão* vai determinar: De todas as pessoas que eu classifiquei como risco cardiaco, quais realmente tinham risco ?

* A *revocação*: De todas as pessoas que chegaram até o médico, com risco cardiaco, quais realmente tinham risco cardiaco. Pois chegaram pessoas que não estavam classificadas com risco e na verdade possuiam.

## Medidas de precisão:

* **Acurácia**: Porcentagem de elementos classificados corretamente (positivos ou negativos). 

    * Representa a porcentagem dos dados que são corretamente classificados pelo modelo

    * $A = \frac{(T_P + T_N)}{(T_P + T_N + F_P + F_N)} $

* **Acurácia por classe**:

    * Calcula-se a média das acurácias individuais para cada classe.

    * Minimiza o problema de desbalanceamento de classe.

        * Atenua o problema de uma base de dados com classes desbalanceadas.

    * Desvantagem: se uma classe possui poucas amostras, aumenta a variância da medida.

## Medidas de avaliação: Log-Loss

Usado quando um classificador retorna uma probabilidade de classificação ("confiança").

$log-loss = - \frac{1}{N} \sum y_i \ log \ p_i + (1 - y_i) \ log(1 - p_i)$

## Medidas de avaliação: Precisão

Define os chamados positivos verdadeiros. Dentre os exemplo classificados como verdadeiros, quanto eram realmente verdadeiros.

* Representa a porcentagem de acertos dentre os elementos classificados como verdadeiro

* $P = \frac{T_P}{T_P + F_P}$

## Medidas de avaliação: Revocação/Sensitividade

Capacidade de recuperação da classe possitiva. Dentre o total de exemplos verdadeiros, quantos foram classificados como verdadeiros.

* Representa a porcentagem dos elementos originalmente verdadeiros que foram classificados corretamente.

* $R =  \frac{T_P}{(T_P + F_N)}$

## Medidas de avaliação: Especifidade

Capacidade de recuperação da classe negativa. Porcentagem de amostras negativas identificadas corretamente sobre o total de amostas negativas.

* $S = \frac{T_N}{T_N + F_P}$

## Medidas de avaliação: F-measure ou F-score

É a média ponderada de precisão e revocação. É muito utilizada em classificação de imagens.

* Representa a média harmônica entre a capacidade de classificar corretamente elementos verdadeiros e a probabilidade de se classificar elementos falsos como verdadeiros.

* $F = 2 * \frac{PRECISAO \ * \ REVOCAÇÃO}{PRECISAO \ + \ REVOCAÇÃO}$
