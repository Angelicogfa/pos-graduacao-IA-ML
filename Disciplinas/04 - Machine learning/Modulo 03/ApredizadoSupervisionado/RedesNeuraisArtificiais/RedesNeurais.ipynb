# Redes neurais artificiais

Método bioinspirado baseado em redes de neurônios artificiais interconectados.

Vantagens:

* Alta acurácia e robusto à bases com erros

* Saida pode ser discreta (classficação) ou contínua (previsão) ou multivalorada.

Críticas:

* Treinamento demorado e sensível a diversos parâmetros tais como topologia de rede, número de neurônios, taxa de apredizado, número de épocas utilizadas.

* Difícil de compreender a função aprendida (pesos).
