# Regressão linear

É o processo de estimar uma linha reta que vai tentar prever uma saída com base em valores de entrada. Por ser um modelo linear, o algoritmo nem sempre consegue prever corretamente pois nem todo problema é linear, ou seja, pode ser separado por uma reta.

## Regulaziração Ride (L2 - quadratica)

* Reduzir o valor absoluto dos parâmetros (magnitude), evitando colinearidade.

    * Diferença absoluta dos valores entre parâmetros.

* Reduz a complexidade do modelo pelo encolhumento do coeficiênte.

* Usa técnica de regulização chamada L2.

## Regularizalçao LASSO (L1 - least absolute shrinkage selector operator)

* Reduz a quantidade de parâmetros aproximando-os do zero absoluto (feature selection, não presente no ridge)

    * Zera os atributos menos significativos, que estão contribuindo pouco.

* Usa uma técnica chamada L1.

* Usada quando se tem uma grande quantidade de atributos.
