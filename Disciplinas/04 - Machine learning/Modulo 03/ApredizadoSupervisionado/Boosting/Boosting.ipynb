# Boosting - Combinação de modelos

Método para melhorar a precisão de qualquer algoritmo de aprendizado.

Funciona criando uma série de datasets tal que um mesmo desempenho modesto sobre essa base baseados pode ser utlizada para construir um previsor de alta precisão.

## Adaboost (Adaptive Boosting)

Adaboost é considerado um dos top 10 do machine learning.

Seja o conjunto de dados:

S = {($x_1, y_1$), ($x_2, Y_2$), ($x_3, y_3$)...($X_n, y_n$)}, onde $x \in X, y \in [-1,1]$ E um modelo de aprendizado fraco $A$.


Para $t = 1,2,3....T$

1. Construir Domínio $D_t$ sobre {$x_1, x_2, x_3..., X_n$}

2. Executar $A$ sobre $D_t$ produzindo $H_T: X \to Y$

3. Seleciona modelo com menor erro

$\epsilon_t = P_{xi} \sim D_i (h_i(x_i) \neq  y_i)$, erro $h_i$ sobre $D_i$ 

Saída: $H = sign(\sum_{t=1} a_t h_t(x))$

* Pode usar qualquer classificador fraco

* É rápido, pois faz apenas uma passada na base a cada iteração

* Mudança de mentálidade: O objetivo é encontrar um classificador que seja marginalmente melhor que adivinhação.
